```python
offer_country_before = df[
    (   df['time_year'] == 2018) & 
        ((df['time_month'] >= 9) & 
        (df['time_month'] <= 12)    )     |
    (   df['time_year'] == 2019) & 
        ((df['time_month'] >= 1) & 
        (df['time_month'] <= 8)    )     

].countries.value_counts()
```

### after epidemic data extract: offer * countries

```python
offer_country_after = df[
    (   df['time_year'] == 2020) & 
        ((df['time_month'] >= 9) & 
        (df['time_month'] <= 12)    )     |
    (   df['time_year'] == 2021) & 
        ((df['time_month'] >= 1) & 
        (df['time_month'] <= 9)    )     

].countries.value_counts()
```

```python
offer_country_after
```

```python
dict_after = offer_country_after.to_dict()
```

```python
dict_after
```

### creat a csv to store the full data which used to visualize

```python
before = pd.DataFrame({'countries':dict_before.keys(),'offers':dict_before.values(),'phase':2018})
before
```

```python
after = pd.DataFrame({'countries':dict_after.keys(),'offers':dict_after.values(),'phase':2020})
after
```

```python
alldata = before.append(after)
alldata
```

```python
alldata.to_csv('offer_by_countries_before_after_covid19.csv')
```

```python
countries = ['美国', '香港','英国', '加拿大','新加坡','荷兰','瑞典','瑞士','丹麦',
           '澳大利亚','意大利']

colors = ['#B71C1C','#AD1457','#7B1FA2','#3F51B5','#42A5F5','#80DEEA','#B2DFDB','#FFCA28','#F57C00','#BF360C','#263238']
#'#B71C1C','#AD1457','#7B1FA2','#5E35B1','#3F51B5','#42A5F5','#4FC3F7','#80DEEA','#B2DFDB','#FFF176','#FFCA28','#F57C00','#BF360C','#4E342E','#263238'
fig, ax = plt.subplots(1, figsize=(10,10))
#ax.set_facecolor('darkgrey')

for i, v in enumerate(countries):
    # get a single country from the list
    temp = alldata[alldata['countries'] == v]
    # plot the lines
    plt.plot(temp.phase, temp.offers, 
             color=colors[i], lw=2.5, 
             marker='o', markersize=5)
    plt.text(temp.phase.values[0]-0.05, 
             temp.offers.values[0], 
             temp.offers.values[0],ha='right')
    
    plt.text(temp.phase.values[1]+0.05, 
             temp.offers.values[1], 
             temp.offers.values[1], va='center', ha='left')
    
    correction = 0
    if v == '美国': correction = 2200
    if v == '香港': correction = 900
    if v == '英国': correction = 1200
    if v == '新加坡': correction = 400
    #if v == '美国': correction = 2200
    plt.text(2019, 
             temp.offers.values[0] - correction, 
             v, color=colors[i],
             va='center', ha='center', fontsize=14)

    
# x limits, x ticks, and y label 
plt.xlim(2017.5,2020.5)
plt.xticks([2018, 2020])
#yticks = plt.yticks()[0]
plt.yticks([])

# grid
#ax.xaxis.grid(color='black', linestyle='solid', which='both', alpha=0.9)
ax.xaxis.grid(color='black', linestyle='solid', 
              which='both', alpha=0.9)
# remove spines
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.title('offer amount of countries(regions)', loc='left', fontsize=20)

#plt.legend(countries, loc='upper left', frameon=False)

plt.show()
```

```python
countries = ['美国', '香港','英国', '加拿大','新加坡','荷兰','澳大利亚','瑞典','瑞士','丹麦',
           '意大利']

colors = ['#B71C1C','#AD1457','#7B1FA2','#3F51B5','#42A5F5','#80DEEA','#B2DFDB','#FFCA28','#F57C00','#BF360C','#263238']
#'#B71C1C','#AD1457','#7B1FA2','#5E35B1','#3F51B5','#42A5F5','#4FC3F7','#80DEEA','#B2DFDB','#FFF176','#FFCA28','#F57C00','#BF360C','#4E342E','#263238'
fig, ax = plt.subplots(1, figsize=(10,15),dpi=100)
#ax.set_facecolor('darkgrey')

for i, v in enumerate(countries):
    # get a single country from the list
    temp = alldata[alldata['countries'] == v]
    # plot the lines
    plt.plot(temp.phase, temp.offers, 
             color=colors[i], lw=2.5, 
             marker='o', markersize=5)
    if i <= 4:
        plt.text(temp.phase.values[0]-0.05, 
                 temp.offers.values[0], 
                 temp.offers.values[0],ha='right')

        plt.text(temp.phase.values[1]+0.05, 
                 temp.offers.values[1], 
                 temp.offers.values[1], va='center', ha='left')

        correction = 0
        if v == '美国': correction = 2200
        if v == '香港': correction = 900
        if v == '英国': correction = 1100
        #if v == '新加坡': correction = 400
        #if v == '美国': correction = 2200
        plt.text(2019, 
                 temp.offers.values[0] - correction, 
                 v, color=colors[i],
                 va='center', ha='center')
    if i == 9:
        plt.text(temp.phase.values[0]-0.05, 
                 temp.offers.values[0], 
                 temp.offers.values[0],ha='right')

        plt.text(temp.phase.values[1]+0.05, 
                 temp.offers.values[1], 
                 temp.offers.values[1], va='center', ha='left')

        correction = 250
        #if v == '美国': correction = 2200
        plt.text(2019, 
                 temp.offers.values[0] - correction, 
                 v, color=colors[i],
                 va='center', ha='center', fontsize=14)

    
# x limits, x ticks, and y label 
plt.xlim(2017.5,2020.5)
plt.xticks([2018, 2020],fontsize=18)
#yticks = plt.yticks()[0]
plt.yticks([],fontsize=18)

# grid
#ax.xaxis.grid(color='black', linestyle='solid', which='both', alpha=0.9)
ax.xaxis.grid(color='black', linestyle='solid', 
              which='both', alpha=0.9)
# remove spines
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.title('Application number in different countries/regions)', loc='center', fontsize=25)

#左上角的图例
plt.legend(countries, frameon=False,fontsize=12,loc=(-0.03,0.63))
#plt.savefig("plot9.jpg")
plt.show()
```
